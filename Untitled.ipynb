```python
# Classification with logistic regression

import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create and train the Logistic Regression model
model = LogisticRegression(max_iter=1000)  # Increased max_iter to ensure convergence
model.fit(x_train, y_train)

# Make predictions on the test set
y_pred = model.predict(x_test)

# Evaluate the model using accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
#clustering with K_MEANS

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans  # Use KMeans instead of k_means


# Load the Iris dataset
iris = load_iris()
X = iris.data

# Perform K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)  # Create KMeans object
kmeans.fit(X)  # Fit the model to the data
labels = kmeans.labels_  # Get cluster assignments

# Visualize the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')  # Scatter plot
plt.xlabel('Sepal Length (cm)')  # X-axis label
plt.ylabel('Sepal Width (cm)')  # Y-axis label
plt.title('K-Means Clustering of Iris Dataset')  # Title
plt.show()  # Display the plot
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error

# Load data
data = pd.read_csv('/content/social_media_engagement.csv', header=0, index_col=0, parse_dates=True)
print(data.head())  # Print the first few rows to inspect the data

# Assuming 'comments' and 'shares' are columns in your data
# and you want to use 'comments' for training and 'shares' for testing

# Split data into training and testing sets
# Adjust the split point (e.g., 0.7) as needed
train_size = int(len(data) * 0.7)
train_data = data['comments'][:train_size]
test_data = data['shares'][train_size:]

# Create and fit the ARIMA model
# The order (5, 1, 0) represents (p, d, q)
# p: Autoregressive order
# d: Differencing order
# q: Moving average order
model = ARIMA(train_data, order=(5, 1, 0))  # Using train_data here
model_fit = model.fit()

# Make predictions
predictions = model_fit.predict(start=len(train_data), end=len(data)-1) # Corrected prediction range


# Evaluate the model
mse = mean_squared_error(test_data, predictions)  # Using test_data here
print(f"Mean Squared Error: {mse:.2f}")

# Plot actual vs predicted values
plt.plot(test_data.index, test_data.values, label='Actual')  # Using index for x-axis
plt.plot(test_data.index, predictions, label='Predicted')    # Using index for x-axis
plt.legend()
plt.show()
```
