```python

import matplotlib.pyplot as plt

# Data
Month = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
Sales = [20000, 25000, 22000, 27000, 30000, 31000, 28000, 29000, 32000, 33000, 34000, 36000]

# Plotting
plt.plot(Month, Sales)
plt.title("Samson's Sample")
plt.xlabel("Months")
plt.ylabel("Sales (in $)")
plt.show()

```
